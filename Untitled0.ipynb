Для начала нам нужно избавиться от пропусков с помощью команды movies.dropna(inplace=True). Эта команда означает удалить все фильмы, у которых не указана прибыль. Суть команды в том, чтобы выбросить пропущенные значения, drop - выкинуть, na - Not a Number/Null - пропущенные или пустые значения. В круглых скобках мы добавили inplace = True. Это означает, что мы хотим перезаписать исходные данные.

```python
movies.dropna(inplace=True)
```
